```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
file_to_load = os.path.join("Resources","purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
#purchase_data.dtypes
```

```python
purchase_data = pd.DataFrame(purchase_data)
purchase_data['SN'] = purchase_data['SN'].astype(str)
purchase_data.dtypes
```

## Player Count

* Display the total number of players

```python
PlayerCount = len(purchase_data['SN'].value_counts())
PlayerCount
```

## Purchasing Analysis (Total)

```python
datacategories = purchase_data.columns
datacategories
purchase_data.head()
```

```python
purchase_data.count()
```

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
UniqueItems = len(purchase_data["Item Name"].value_counts())

Average = purchase_data['Price'].mean()

NumberPurchases = purchase_data['Price'].count()

TotalRevenue = purchase_data['Price'].sum()
TotalRevenue
```

```python
Summary = purchase_data.describe()

data = {'Number of Unique Items': [UniqueItems], 'Average Purchase Price': [Average], 'Total Number of Purchases': [NumberPurchases], 'Total Revenue': [TotalRevenue]}
PurchasingAnalysis_Total = pd.DataFrame.from_dict(data) 
PurchasingAnalysis_Total['Average Purchase Price'] = PurchasingAnalysis_Total['Average Purchase Price'].map('${:,.2f}'.format)
PurchasingAnalysis_Total['Total Revenue'] = PurchasingAnalysis_Total['Total Revenue'].map('${:,.2f}'.format)
PurchasingAnalysis_Total
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
GenderGroup = purchase_data.groupby(["Gender"], as_index=True)['SN'].nunique()
GenderGroup = pd.DataFrame(GenderGroup)
#GenderGroup['SN'] = GenderGroup[]

GenderGroup.columns = ['Player Count']

GenderDistribution = [x/PlayerCount for x in GenderGroup["Player Count"]]

GenderDistribution = pd.DataFrame({
    'Player Count': GenderGroup['Player Count'],
    '% of Purchasers': GenderDistribution
    }
)

#genindex = ['Female', 'Male', 'Other/Non-Disclosed']
#GenderDistribution.set_index(genindex)
#GenderGroup = pd.merge(GenderGroup, GenderDistribution, how='left', left_index=True, right_index=True)
#GenderGroup.columns = ['Player Count', '% of Purchasers']
GenderDistribution["Player Count"] = GenderDistribution['Player Count'].map('{:}'.format)
GenderDistribution['% of Purchasers'] = GenderDistribution['% of Purchasers'].map('{:.0%}'.format)
GenderDistribution
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
PurchaseAnalysis_Gender = purchase_data.groupby(['Gender']).agg(
    {
         'Purchase ID': "count",
         'Price': ["mean", "sum"]  
    }
)

PurchaseAnalysis_Gender.columns = ['purchase_count', 'average_cost', 'total_value']
GenderSNgroup = purchase_data.groupby(['Gender', 'SN']).sum() 
# I disagree with the way it's phrased, but this is what they probably are looking for.


AvgPurchasePerPerson = pd.DataFrame(GenderSNgroup.Price)
AvgPurchasePerPerson = AvgPurchasePerPerson.groupby(['Gender']).mean()
AvgPurchasePerPerson.columns = ['avg_purchase_total_per_person']

PurchaseAnalysis_Gender = pd.merge(PurchaseAnalysis_Gender, AvgPurchasePerPerson, how='left', left_index=True, right_index=True)

#AvgPurchasePerPerson




#PurchaseAnalysis_Gender['purchase_count'] = PurchaseAnalysis_Gender['purchase_count'].map('{:}'.format)
PurchaseAnalysis_Gender['average_cost'] = PurchaseAnalysis_Gender['average_cost'].map('${:.2f}'.format)
PurchaseAnalysis_Gender['total_value'] = PurchaseAnalysis_Gender['total_value'].map('${:.2f}'.format)
PurchaseAnalysis_Gender['avg_purchase_total_per_person'] = PurchaseAnalysis_Gender['avg_purchase_total_per_person'].map('${:.2f}'.format)

PurchaseAnalysis_Gender.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]
PurchaseAnalysis_Gender
```

## Age Demographics

```python
SortedAges = pd.DataFrame(purchase_data.sort_values('Age', ascending=True))
AgeMin = SortedAges.Age.min()
AgeMax = SortedAges.Age.max()
AgeRange = AgeMax - AgeMin
AgeRange
AgeMin
BinCuts = pd.interval_range(start = AgeMin-1, freq = 4, end = AgeMax+3, closed='right')
BinDF = pd.DataFrame(BinCuts)
BinDF.columns = ['AgeBin']

#SortedAges = pd.DataFrame()
Bin = pd.DataFrame(pd.cut(SortedAges.Age, bins=BinCuts, include_lowest=True))
Bin.columns = ['AgeBin']
#purchase_data.append(AgeBinsData)
#BinCuts
#SortedAges['Bin'] = BinCuts

#purchase_data.describe()
SortedAges = pd.merge(SortedAges, Bin, how='left', left_index=True, right_index=True )
SortedAges
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
PurchaseAnalysis_Age = SortedAges.groupby(['AgeBin']).agg(
    {
         'Purchase ID': "count",
         'Price': ["mean", "sum"]  
    }
)

PurchaseAnalysis_Age.columns = ['purchase_count', 'average_cost', 'total_value']
AgeSNgroup = SortedAges.groupby(['SN']).agg(
    {
        'Price': 'sum',
        'AgeBin': 'first'
    }
)

AgeSNgroup


AvgPurchasePerPersonAge = pd.DataFrame(AgeSNgroup)

AvgPurchasePerPersonAge = AvgPurchasePerPersonAge.groupby(['AgeBin']).mean()
AvgPurchasePerPersonAge.columns = ['avg_purchase_total_per_person']

PurchaseAnalysis_Age = pd.merge(PurchaseAnalysis_Age, AvgPurchasePerPersonAge, how='left', left_index=True, right_index=True)

#AvgPurchasePerPerson




#PurchaseAnalysis_Gender['purchase_count'] = PurchaseAnalysis_Gender['purchase_count'].map('{:}'.format)
PurchaseAnalysis_Age['average_cost'] = PurchaseAnalysis_Age['average_cost'].map('${:.2f}'.format)
PurchaseAnalysis_Age['total_value'] = PurchaseAnalysis_Age['total_value'].map('${:.2f}'.format)
PurchaseAnalysis_Age['avg_purchase_total_per_person'] = PurchaseAnalysis_Age['avg_purchase_total_per_person'].map('${:.2f}'.format)

PurchaseAnalysis_Age.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]
PurchaseAnalysis_Age.index.names = ['Age Bin']


PurchaseAnalysis_Age
#AvgPurchasePerPersonAge

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SortedAges
SpenderSN = SortedAges.SN.value_counts()
TopSpenders = SortedAges.groupby('SN', as_index=False).agg(
    {'Item Name': 'count',
     'Price': ['mean','sum']
        
    })
TopSpenders.columns = ['SN', 'Purchase Count', 'Average Purchase Price', 'Total Purchase Value']

TopSpenders = TopSpenders.sort_values(by=['Total Purchase Value'], ascending=False)

TopSpenders['Average Purchase Price'] = TopSpenders['Average Purchase Price'].map('${:.2f}'.format)
TopSpenders['Total Purchase Value'] = TopSpenders['Total Purchase Value'].map('${:.2f}'.format)

TopSpenders = TopSpenders.iloc[:5]
TopSpenders
```

## Most Popular Items

* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results

```python
ItemsOnly = pd.DataFrame({
    'Item ID': SortedAges['Item ID'],
    'Item Name': SortedAges['Item Name'],
    'Item Price': SortedAges['Price']
    }
)

Popular = ItemsOnly.groupby(['Item ID', 'Item Name'], as_index=False).agg({
    'Item Price': ['mean', 'sum']
})

Popular.columns = ['Item ID', 'Item Name', 'Item Price', 'Total Purchase Value']

PopCount = ItemsOnly['Item Name'].value_counts()
PopCount = pd.DataFrame(PopCount)
PopCount = PopCount.reset_index()
PopCount.columns = ['Item Name','Purchase Count']
PopularItems = pd.merge(Popular, PopCount, how='left', left_on='Item Name', right_on='Item Name')

'''
SpenderSN = SortedAges.SN.value_counts()
TopSpenders = SortedAges.groupby('SN', as_index=False).agg(
    {'Item Name': 'count',
     'Price': ['mean','sum']
        
    })
TopSpenders.columns = ['SN', 'Purchase Count', 'Average Purchase Price', 'Total Purchase Value']

TopSpenders = TopSpenders.sort_values(by=['Total Purchase Value'], ascending=False)

TopSpenders['Average Purchase Price'] = TopSpenders['Average Purchase Price'].map('${:.2f}'.format)
TopSpenders['Total Purchase Value'] = TopSpenders['Total Purchase Value'].map('${:.2f}'.format)

TopSpenders = TopSpenders.iloc[:5]
TopSpenders
'''
#Popular.columns = ['Item ID', 'Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']

PopularItems = PopularItems.sort_values(by=['Purchase Count'], ascending=False)
PopularItems = PopularItems[['Item ID', 'Item Name','Purchase Count', 'Item Price','Total Purchase Value']]
PopularItemsTop5 = PopularItems
PopularItemsTop5 = PopularItemsTop5.iloc[:5]

PopularItemsTop5['Item Price'] = PopularItemsTop5['Item Price'].map('${:.2f}'.format)
PopularItemsTop5['Total Purchase Value'] = PopularItemsTop5['Total Purchase Value'].map('${:.2f}'.format)



PopularItemsTop5
```
